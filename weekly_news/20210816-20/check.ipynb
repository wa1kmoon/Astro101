```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np
import pandas as pd
```

```python
sd=pd.read_csv('./Seeing_Data.csv', low_memory=False, header=None)
```

```python
print(sd)
```

```python
jd=np.array(sd.loc[:,2]).astype('float')
seeing=np.array(sd.loc[:,5]).astype('float')
```

```python
fig1=plt.figure(figsize=(25,7))
plt.scatter(jd,seeing)
y_loc=MultipleLocator(0.5)
x_loc=MultipleLocator(90)
ax=plt.gca()
ax.yaxis.set_major_locator(y_loc)
ax.xaxis.set_major_locator(x_loc)
```

```python
fig2=plt.figure(figsize=(10,7))
plt.hist(seeing, bins=60)
ax=plt.gca()
ax.xaxis.set_major_locator(y_loc)
```
